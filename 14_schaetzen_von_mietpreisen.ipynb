# Input und Outputdaten normalisieren

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_input = MinMaxScaler(feature_range=(0, 1))
scaled_input_data = scaler_input.fit_transform(data_input)
```

```python
scaled_input_data
```

```python
scaled_input_data.shape
```

```python
scaler_output = MinMaxScaler(feature_range=(0, 1))
scaled_output_data = scaler_output.fit_transform(data_output)
```

```python
scaled_output_data
```

```python
scaled_output_data.shape
```

# Trainings- und Testdaten definieren

```python
anz_gefilterter_bsp = scaled_input_data.shape[0]
```

```python
anz_train_bsp = int(0.8*anz_gefilterter_bsp)
anz_train_bsp
```

```python
x_train = scaled_input_data[0:anz_train_bsp]
y_train = scaled_output_data[0:anz_train_bsp]

x_test  = scaled_input_data[anz_train_bsp:]
y_test  = scaled_output_data[anz_train_bsp:]
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# MLP vorbereiten

```python
anz_inputs = x_train.shape[1]
```

```python
anz_inputs
```

```python
import tensorflow as tf
from tensorflow import keras

tf.keras.backend.clear_session()

model = keras.Sequential()
model.add(keras.layers.Dense(10000,
                             activation="relu",
                             input_shape=(anz_inputs,)))
model.add(keras.layers.Dense(1,
                             activation="linear"))
```

```python
model.compile(optimizer='sgd',               
              loss=tf.keras.losses.MeanSquaredError())              
```

```python
model.summary()
```

# MLP trainieren

```python
x_train.shape
```

```python
y_train.shape
```

```python
history = model.fit(x_train,
                    y_train,
                    validation_split = 0.8,
                    epochs=200)
```

# Plotten der Fehlerkurve

```python
history.history
```

```python
import matplotlib.pyplot as plt
y1 = history.history["loss"]
y2 = history.history["val_loss"]
plt.plot(y1, color="red")
plt.plot(y2, color="green")
plt.xlabel("Trainingsepoche")
plt.ylabel("Fehler")
plt.title("Fehlerkurven")
plt.show()
```
