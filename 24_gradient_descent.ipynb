# Gradient descent for a function f:R->R

## Define an example function

```python
import numpy as np

def f(x):
    return x*np.cos(x)

xs = np.arange(-10.0,+10.0, 0.1)
ys = [f(x) for x in xs]
```

```python
import matplotlib.pyplot as plt

plt.scatter(xs,ys, marker=".", color="gray")
plt.grid()
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.show()
```

## Derivative of this function

```python
def df(x):
    """
    f(x)  = x*cos(x)
    f'(x) = 1*cos(x) + x*-sin(x)
    """
    return np.cos(x) - x*np.sin(x)
```

```python
dfs = [df(x) for x in xs]
```

```python
plt.scatter(xs,ys,  marker=".", color="gray")
plt.scatter(xs,dfs, marker="+", color="green", s=5)
plt.grid()
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.show()
```

## Gradient descent

```python
def gradient_descent(startx, step_size=0.1, nr_steps=5):
    
    posx = startx
    visitedposx = [posx]
    
    for step_nr in range(nr_steps):
        
        gradient = df(posx)
        posx = posx - (gradient*step_size)
        
        visitedposx.append(posx)
        
    return visitedposx
```

```python
visitedposx = gradient_descent( startx=6.0, step_size=0.05, nr_steps=5 )
visitedposx
```

```python
def show_gradient_descent_trajectory(visitedposx):
    
    visitedposy = [f(x) for x in visitedposx]

    plt.figure(figsize=(9,7))
    plt.scatter(xs,ys, marker=".", color="gray")
    plt.scatter(visitedposx,visitedposy, marker="o", color="red", s=20)
    plt.grid()
    plt.xlabel("x")
    plt.ylabel("y=f(x)")
    plt.show()
```

```python
show_gradient_descent_trajectory(visitedposx)
```

```python
visitedposx = gradient_descent( startx=-2.0, step_size=0.05, nr_steps=15 )
show_gradient_descent_trajectory(visitedposx)
```
