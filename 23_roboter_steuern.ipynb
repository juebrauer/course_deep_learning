Hier ein paar Vorabinfos:


    Landing pad is always at coordinates (0,0).
    Coordinates are the first two numbers in state vector.
    Reward for moving from the top of the screen to landing pad and zero speed is about 100..140 points.
    If lander moves away from landing pad it loses reward back.
    Episode finishes if the lander crashes or comes to rest, receiving additional -100 or +100 points.
    Each leg ground contact is +10. Firing main engine is -0.3 points each frame.
    Solved is 200 points.
    Landing outside landing pad is possible.
    Fuel is infinite, so an agent can learn to fly and then land on its first attempt.
    Four discrete actions available:
    0 - do nothing
    1 - fire right orientation engine
    2 - fire main (down) engine
    3 - fire left orientation engine

Eine Beschreibung des Zustandsvektors (observation) findet sich hier:
https://arxiv.org/pdf/2011.11850.pdf

    x coordinate of the lander
    y coordinate of the lander
    vx, the horizontal velocity
    vy, the vertical velocity
    θ, the orientation in space
    vθ, the angular velocity
    Left leg touching the ground (Boolean)
    Right leg touching the ground (Boolean)


done=True sobald einer der beiden Landefüsse Kontakt hat

# Lander selber steuern

```python
ordner_name = "flugdaten_gerade_landen"
```

```python
# Code ist stark inspiriert von:
# https://github.com/StanfordASL/ALPaCA/blob/master/gym/examples/agents/keyboard_agent.py

import sys, gym, time
import numpy as np
import pandas

env = gym.make("LunarLander-v2")

human_agent_action = 0
human_wants_restart = False
human_sets_pause = False
human_wants_to_save_simulation = False
human_wants_to_exit = False

def taste_gedrueckt(key, mod):
    """
    Eine Taste wurde runter gedrückt.
    Welche? Taste auf Aktion abbilden.
    """
    global human_agent_action, human_wants_restart, human_sets_pause, \
           human_wants_to_save_simulation, human_wants_to_exit
       
        
    # Welche Taste wurde gedrückt?
    
    # p Taste? --> Pause Start bzw. Pause Ende    
    if key==112: human_sets_pause = not human_sets_pause
    
    # Return-Taste? --> Simulation speichern, Neue Simulation starten
    if key==0xff0d:        
        human_wants_restart = True
        human_wants_to_save_simulation = True
        
    # Space-Taste? --> Simulation nicht speichern, Neue Simulation starten    
    if key==32:
        human_wants_restart = True
        human_wants_to_save_simulation = False
        
    # Exit-Taste?
    if key==120:
        human_wants_to_exit = True        
    
    # Steuerdüse?
    a=0
    if key==97: # Taste a
        a = 3
    if key==100: # Taste d
        a = 1
    if key==115: # Taste s
        a = 2
    human_agent_action = a

def taste_losgelassen(key, mod):
    """
    Die Taste wurde wieder los gelassen -> keine Aktion
    """
    global human_agent_action
    human_agent_action = 0
    
env.render()
env.unwrapped.viewer.window.on_key_press = taste_gedrueckt
env.unwrapped.viewer.window.on_key_release = taste_losgelassen

def eine_landungs_simulation(env):
    global human_agent_action, human_wants_restart, human_sets_pause, human_wants_to_exit
    human_wants_restart = False
    obser = env.reset()
    
    tabelle = pandas.DataFrame(columns=["x", "y","vx","vy","orient","vorient","left","right","action"])
    
    simu_step = 0
    while True:
        
        a = human_agent_action
            
        obser, r, done, info = env.step(a)
        
        # eine weitere Zeile zur Pandas Tabelle hinzufügen
        s = pandas.Series({ "x"       : obser[0],
                            "y"       : obser[1],
                            "vx"      : obser[2],
                            "vy"      : obser[3],
                            "orient"  : obser[4],
                            "vorient" : obser[5],
                            "left"    : obser[6],
                            "right"   : obser[7],
                            "action"  : a,
                          })
        s.name = simu_step              
        tabelle = tabelle.append(s)
        
        """
        # Landung muss unter bestimmten Winkel erfolgen
        if done:            
            vel_x = obser[2]
            vel_y = obser[3]
            orient_rad = obser[4]
            orient_grad = orient_rad * (180/np.pi)
                        
            if abs(orient_grad)<15.0:
                print("Landeorientierung ok")
        """               
            
        window_still_open = env.render()
              
            
        # Möchte der Benutzer eine neue Simulation starten?
        if human_wants_restart or human_wants_to_exit:          
            return tabelle
            
        # Pause?
        while human_sets_pause:
            env.render()
            time.sleep(0.1)
            
        # Nicht so schnell! Warten... [in Sekunden]
        time.sleep(0.1)
        
        simu_step += 1
       
    

        
        
anleitung = """
Willkommen zur Raumschifflandung!

a = Steuerdüse links
s = Steuerdüse unten
d = Steuerdüse rechts

p = Pause start/stop
x = Exit

Return = Simulation       als Trainingsbeispiel speichern, gehe zur nächsten
Space  = Simulation NICHT als Trainingsbeispiel speichern, gehe zur nächsten
"""
print(anleitung)

import os
try:    
    os.makedirs( ordner_name )
    print( f"Verzeichnis {ordner_name} erstellt." )
except FileExistsError:
    print( f"Verzeichnis {ordner_name} gibt es schon!" )    
datei_nr = 1

simulation_nr = 1
while True:
    print( f"Simulation #{simulation_nr}")    
    tabelle = eine_landungs_simulation( env )
    
    if human_wants_to_exit:
        break        
    
    if human_wants_to_save_simulation:
        dateiname = f"{ordner_name}/{datei_nr:04}.csv"
        print( f"Simulation/Flugdaten werden gespeichert in {dateiname}" )        
        tabelle.to_csv( dateiname )
        datei_nr += 1
    else:
        print("Simulation nicht speichern.")
        
    simulation_nr +=1
    
print("Programmende!")
```

# Neuronales Netz vorbereiten und trainieren

```python
import tensorflow as tf
tf.__version__
```

```python
anz_inputs = x_train.shape[1]
anz_inputs
```

```python
anz_outputs = y_train.shape[1]
anz_outputs
```

```python
from tensorflow import keras

tf.keras.backend.clear_session()

model = keras.Sequential()
model.add(keras.layers.Dense(20,
                             activation="tanh",
                             input_shape=(anz_inputs,),
                             name="HiddenLayer1"))
model.add(keras.layers.Dense(anz_outputs,
                             activation="linear",
                             name="OutputLayer"))

model.compile(optimizer='sgd',
              loss=keras.losses.MeanSquaredError())

model.summary()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
history = model.fit(x_train,
                    y_train,
                    validation_split=0.1,
                    epochs=50)
```

```python
import matplotlib.pyplot as plt
y1 = history.history["loss"][1:]
y2 = history.history["val_loss"][1:]
plt.plot(y1, color="red", label="loss")
plt.plot(y2, color="green", label="val_loss")
plt.xlabel("Trainingsepoche")
plt.ylabel("Fehler")
plt.title("Fehlerkurven")
plt.legend()
plt.show()
```

# Neuronales Netz auf Testdaten testen

```python
preds = model.predict( x_test )
```

```python
preds[:10]
```

```python
N = y_test.shape[0]

correct = 0
for test_nr in range(N):
    
    human_action = np.argmax(y_test[test_nr])
    predicted_action = np.argmax(preds[test_nr])
    
    if predicted_action == human_action:
        correct +=1
    
    print( f"test situation #{test_nr}: human action={human_action} vs. predicted action={predicted_action}" )
    
print( f"actions correct: {(correct/N)*100.0}")
```

# Neuronales Netz im Simulator testen

```python
import sys, gym, time
import numpy as np
import pandas

env = gym.make("LunarLander-v2")

human_agent_action = 0
human_wants_restart = False
human_sets_pause = False
human_wants_to_save_simulation = False
human_wants_to_exit = False

def taste_gedrueckt(key, mod):
    """
    Eine Taste wurde runter gedrückt.
    Welche? Taste auf Aktion abbilden.
    """
    global human_agent_action, human_wants_restart, human_sets_pause, \
           human_wants_to_save_simulation, human_wants_to_exit
    
    # Exit-Taste?
    if key==120:
        human_wants_to_exit = True        
    
def taste_losgelassen(key, mod):
    """
    Die Taste wurde wieder los gelassen -> keine Aktion
    """
    pass
    
env.render()
env.unwrapped.viewer.window.on_key_press = taste_gedrueckt
env.unwrapped.viewer.window.on_key_release = taste_losgelassen

def eine_landungs_simulation(env):
    global human_wants_to_exit
    
    obser = env.reset()
        
    simu_step = 0
    while True:
        
        x  = obser[0]
        y  = obser[1]
        vx = obser[2]
        vy = obser[3]
        orient = obser[4]
        vorient = obser[5]
        
        input_vec = np.array( [x,y,vx,vy,orient,vorient] )
        #input_vec = np.array( [vx,vy,orient,vorient] )
        input_matrix_scaled = scaler_input.transform( input_vec.reshape(1,-1) )
        
        # frage das NN was zu tun ist!
        preds = model.predict( input_matrix_scaled )        
        a = np.argmax( preds[0] )
        #print( f"vorgeschlagene Aktion: {a}" )
            
        obser, r, done, info = env.step(a)
                    
        window_still_open = env.render()
              
        # Exit?
        if human_wants_to_exit:          
            return
        
        # Gelandet?
        if done:
            return
            
        # Nicht so schnell! Warten... [in Sekunden]
        time.sleep(0.05)
        
        simu_step += 1
       
    

        
        
anleitung = """
Das Neuronales Netz steuert jetzt das Raumschiff!
x = Exit
"""
print(anleitung)

simulation_nr = 1
while True:
    print( f"Simulation #{simulation_nr}")    
    eine_landungs_simulation( env )
    
    if human_wants_to_exit:
        break        
        
    simulation_nr +=1
    
print("Programmende!")
```
